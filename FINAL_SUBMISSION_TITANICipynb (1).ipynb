IMPORTING THE NECESSARY LIBRARIES

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

READ THE CSV FILE

```python
titanic = pd.read_csv("train.csv")
titanic
```

TO CHECK NULL VALUES

```python
titanic.isnull().sum()
```

**DATA CLEANING**

```python
#Dropping the irrelevant Columns
tit = titanic.drop(labels=["Cabin","Name","Ticket"], axis=1)
tit
```

```python
tit["Pclass"].unique()
```

```python
#Fill the null values of 'Age' column with the mean value
tit["Age"].fillna(tit["Age"].mean(),inplace = True)
```

```python
#Drooping the Rows with null values
tit.loc[tit['Embarked'].isnull()==True]
```

```python
tit = tit.drop(61)
tit = tit.drop(829).reset_index(drop=True)
```

```python
tit.isnull().sum()
```

```python
#Replacing the male and female genders as 0 and 1
tit.replace(['male','female'],[0,1],inplace = True)
tit
```

```python
#Replacing the Embarked class values with 0,1,2
tit.replace(["Q","S","C"],[0,1,2],inplace = True)
tit
```

```python
tit.info()
```

SEPARATING DATA FOR TRAINING AND TESTING

```python
x = tit.drop(["Survived"],axis = 1)
y = tit['Survived']
```

```python
x
```

```python
y
```

```python
x_train,x_val,y_train,y_val = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

LOGISTIC REGRESSION

```python
model = LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)


```

```python
predictions = model.predict(x_val)
accuracy_score(predictions,y_val)
```

```python
test = pd.read_csv("test.csv")
```

```python
test = test.drop(labels=["Cabin","Name","Ticket"], axis=1)
```

```python
test["Age"].fillna(test["Age"].mean(),inplace = True)
test['Fare'].fillna(test['Fare'].mean(),inplace = True)
test.replace(['male','female'],[0,1],inplace = True)
test.replace(["Q","S","C"],[0,1,2],inplace = True)
```

```python
test.isnull().sum()
```

```python
test_predictions = model.predict(test)
test_ids = test["PassengerId"]
```

```python
df = pd.DataFrame({"PassengerId":test_ids.values,"Survived":test_predictions})
df.to_csv("submission_Kaggle.csv",index=False)
```
